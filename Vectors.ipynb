```python
import tensorflow as tf
```

```python
vector1 = tf.constant([1, 2, 3])
vector2 = tf.constant([4, 5, 6])
```

```python
addition = tf.add(vector1, vector2)
print("Vector Addition Result:", addition.numpy())
```

```python
substraction = tf.subtract(vector1, vector2)
print("Vector Substraction Result:", substraction.numpy())
```

```python
multiplication = tf.multiply(vector1, vector2)
print("Vector Multiplication Result:", multiplication.numpy())
```

```python
division = tf.divide(vector1, vector2)
print("Vector Division Result:", division.numpy())
```
